```python
import numpy as np
import numpy.fft as npft
import scipy.io.wavfile as wavfile


#On importe le ficher son √† restaurer sous forme de vecteur
fs , y = wavfile.read('y.wav')

#On importe l'op√©rateur de convolution permettant de restaurer la musique
h = np.loadtxt('h.txt')

#On calcule ^y=P*y et ^h=P*h √† l'aide de la Transformation de Fourrier Rapide ( fft)
hchap= npft.fft(h)
ychap= npft.fft(y)

#On cr√©er le vecteur des ^y/^h
achap=ychap/hchap

#A l'aide de la fonction ifft on calcule a √† partir du vecteur achap
a=npft.ifft(achap)

#On traduit a en un ficher musique en s'assurant que les coordonn√©es du vecteur sont bien entre -1 et 1 
#√† l'aide de la formule y=(1.1*m)
m=np.max(np.abs(a))
a=a/(1.1*m)
a=a.real
wavfile.write('restored.wav',fs,a)


#Groupe projet mod√© : Akim, Zakari, Guillaume
```
